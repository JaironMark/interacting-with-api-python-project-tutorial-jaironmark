```python
spotify = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials(client_id=client_id,client_secret=client_secret))
results = spotify.artist_top_tracks(BigX)

def traqueador():
    tracks = []  # Mover la inicialización de la lista fuera del bucle
    for track in results['tracks'][:10]:
        # Convertir la duración de milisegundos a minutos y segundos
        duration_ms = track["duration_ms"]
        minutes = duration_ms // 60000  # Obtener los minutos
        seconds = (duration_ms % 60000) // 1000  # Obtener los segundos restantes
        
        track_data = {
            'track_name': track["name"],
            'popularity': track["popularity"],
            'duration': f"{minutes}m {seconds}s",
            'album_name': track["album"]["name"],
            "release_date": track["album"]["release_date"]
        }
        tracks.append(track_data)  # Agregar la canción a la lista

    return tracks  # Ahora imprimir la lista correctamente
```

```python
temitas = pd.DataFrame.from_records(traqueador())
temitas.head(3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear el gráfico de dispersión con marcadores personalizados
sns.scatterplot(
    data=temitas,              # DataFrame con los datos
    x='popularity',            # Eje X: Popularidad
    y='duration',           # Eje Y: Duración en milisegundos
    hue='track_name',          # Color de los puntos según el nombre del track
    marker=(4,1,0),                # Marcador en forma de estrella
    s=300,                     # Tamaño de los marcadores
    palette='plasma',         # Paleta de colores para los puntos
    edgecolor='black',         # Borde de los marcadores en verde
    linewidth=1                # Grosor del borde
)

# Personalizar el gráfico
plt.title('Popularidad vs Duración de las Canciones')
plt.xlabel('Popularidad')
plt.ylabel('Duración minutos')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Mover la leyenda fuera del gráfico
plt.show()
```

```python
import os
import pandas as pd
import seaborn as sns
from dotenv import load_dotenv
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials


# load the .env file variables
load_dotenv()

client_id = os.environ.get("CLIENT_ID")
client_secret = os.environ.get("CLIENT_SECRET")


BigX = "6qxpnaukVayrQn6ViNvu9I"

credenciales = spotipy.Spotify(auth_manager = SpotifyClientCredentials(client_id = client_id,client_secret = client_secret))

spotify = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials(client_id=client_id,client_secret=client_secret))
results = spotify.artist_top_tracks(BigX)

def traqueador():
    tracks = []  # Mover la inicialización de la lista fuera del bucle
    for track in results['tracks'][:10]:
        # Convertir la duración de milisegundos a minutos y segundos
        duration_ms = track["duration_ms"]
        minutes = duration_ms // 60000  # Obtener los minutos
        seconds = (duration_ms % 60000) // 1000  # Obtener los segundos restantes
        
        track_data = {
            'track_name': track["name"],
            'popularity': track["popularity"],
            'duration': f"{minutes}m {seconds}s",
            'album_name': track["album"]["name"],
            "release_date": track["album"]["release_date"]
        }
        tracks.append(track_data)  # Agregar la canción a la lista

    return tracks  # Ahora imprimir la lista correctamente

temitas = pd.DataFrame.from_records(traqueador())


#print(results)
# temitas.head(3)
print(temitas)


import seaborn as sns
import matplotlib.pyplot as plt

# Crear el gráfico de dispersión con marcadores personalizados
sns.scatterplot(
    data=temitas,              # DataFrame con los datos
    x='popularity',            # Eje X: Popularidad
    y='duration',           # Eje Y: Duración en milisegundos
    hue='track_name',          # Color de los puntos según el nombre del track
    marker=(4,1,0),                # Marcador en forma de estrella
    s=100,                     # Tamaño de los marcadores
    palette='inferno',         # Paleta de colores para los puntos
    edgecolor='green',         # Borde de los marcadores en verde
    linewidth=1                # Grosor del borde
)

# Personalizar el gráfico
plt.title('Popularidad vs Duración de las Canciones')
plt.xlabel('Popularidad')
plt.ylabel('Duración minutos')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Mover la leyenda fuera del gráfico
plt.show()
```
